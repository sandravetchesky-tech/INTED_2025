```python
numeros = []

# Ler 15 números do usuário
for i in range(15):
    n = int(input(f"Digite o {i+1}º número: "))
    numeros.append(n)

# Remover repetidos mantendo a ordem
lista_sem_repetidos = []
for num in numeros:
    if num not in lista_sem_repetidos:
        lista_sem_repetidos.append(num)

print("Lista original:", numeros)
print("Sem repetidos:", lista_sem_repetidos)
```

```python
nomes = []
idades = []

# Ler 10 nomes
for i in range(10):
    nome = input(f"Digite o {i+1}º nome: ")
    nomes.append(nome)

# Ler 10 idades
for i in range(10):
    idade = int(input(f"Digite a idade de {nomes[i]}: "))
    idades.append(idade)

# Encontrar a maior idade
maior_idade = max(idades)

# Encontrar o índice da maior idade
indice = idades.index(maior_idade)

# Exibir o nome da pessoa mais velha
print("A pessoa mais velha é:", nomes[indice])
```

```python
numeros = []

# Ler 10 números
for i in range(10):
    n = float(input(f"Digite o {i+1}º número: "))
    numeros.append(n)

# Processar a lista
for i in range(10):
    if numeros[i] % 2 == 0:       # número par
        numeros[i] = numeros[i] / 2
    else:                         # número ímpar
        numeros[i] = numeros[i] / 3

# Exibir resultado
print("Lista modificada:")
print(numeros)
```

```python
# Função para verificar se um número é primo
def eh_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Ler 10 números inteiros
numeros = []
for i in range(10):
    n = int(input(f"Digite o {i+1}º número: "))
    numeros.append(n)

# Criar lista apenas com os números primos
primos = []
for num in numeros:
    if eh_primo(num):
        primos.append(num)

print("Lista original:", numeros)
print("Números primos:", primos)
```

```python
m = [[float(input()) for _ in range(3)] for _ in range(3)]

det = (
    m[0][0]*m[1][1]*m[2][2] +
    m[0][1]*m[1][2]*m[2][0] +
    m[0][2]*m[1][0]*m[2][1] -
    m[0][2]*m[1][1]*m[2][0] -
    m[0][0]*m[1][2]*m[2][1] -
    m[0][1]*m[1][0]*m[2][2]
)

print("Determinante:", det)
```

```python
produtos = {}

for _ in range(3):
    nome = input("Produto: ")
    precos = [float(input("Preço: ")) for _ in range(3)]
    produtos[nome] = precos

maior_var = None
produto_var = ""

for p, valores in produtos.items():
    variacao = max(valores) - min(valores)
    if maior_var is None or variacao > maior_var:
        maior_var = variacao
        produto_var = p

print("Maior variação:", produto_var, maior_var)
```
