```python
salario = float(input("Digite o salário atual: R$ "))

if salario <= 2000:
    aumento = salario * 0.10
elif salario <= 5000:
    aumento = salario * 0.07
else:
    aumento = salario * 0.05

novo_salario = salario + aumento
print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
total = 0
contador = 0

while True:
    idade = int(input("Digite a idade (negativo para encerrar): "))
    if idade < 0:
        break
    total += idade
    contador += 1

if contador > 0:
    media = total / contador
    print(f"Média das idades: {media:.2f}")
else:
    print("Nenhuma idade válida foi digitada.")
```

```python
tarefas = []

for i in range(4):
    tarefa = input(f"Digite o nome da tarefa {i+1}: ")
    tarefas.append(tarefa)

maior_tarefa = tarefas[0]

for tarefa in tarefas[1:]:
    if len(tarefa) > len(maior_tarefa):
        maior_tarefa = tarefa

print(f"A tarefa com o nome mais longo é: {maior_tarefa}")
```

```python
positivos = 0
negativos = 0
zeros = 0

for i in range(10):
    numero = float(input(f"Digite o {i+1}º número: "))
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Números positivos: {positivos}")
print(f"Números negativos: {negativos}")
print(f"Números iguais a zero: {zeros}")
```

```python
senha_correta = "python123"

for tentativa in range(1, 4):
    senha = input(f"Tentativa {tentativa}: Digite a senha: ")
    if senha == senha_correta:
        print("Senha correta! Bem-vindo(a)!")
        break
    else:
        print("Senha incorreta.")
else:
    print("Número máximo de tentativas atingido. Acesso bloqueado.")
```
